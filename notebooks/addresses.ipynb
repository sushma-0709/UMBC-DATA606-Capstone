```python
from snowflake.snowpark import Session
import tomllib
import os
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python

with open("/workspaces/UMBC-DATA606-Capstone/snowflakecli.toml", "rb") as f:
    
    config = tomllib.load(f)  # Python >=3.11

# Get litecoin connection details
litecoin_config = config["connections"]["litecoin"]

# Replace placeholders with environment variable values
conn_params = {k: os.path.expandvars(v) for k, v in litecoin_config.items()}

# Create Snowpark session
session = Session.builder.configs(conn_params).create()

# Test connection
print("Current Snowflake Role:", session.sql("SELECT CURRENT_ROLE()").collect()[0][0])
```
